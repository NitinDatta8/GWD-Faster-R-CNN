### 2.2.1 Start Finetuning here

You can see all options here : https://github.com/ratthachat/wheatdetection/blob/develop/config/defaults.py

Note that 

- load previous checkpoint using `SOLVER.TRAIN_CHECKPOINT True`

- Remember `last-checkpoint.bin` must already inside the `OUTPUT_DIR` 

- I guess we should reduce `base_lr` / `momentum` to be small for finetuning, so I reduce momentum from 0.9 to 0.5 ... `base_lr` is automatically decayed for each epoch, so we set it to the default value of 0.005  

- Default implementation use CosineScheduler. If you don't need CosineScehduler, just simply set `WARMUP_EPOCHS` to be higher than `MAX_EPOCHS` ... (In the code, if `epoch < WARMUP_EPOCHS` , no scheduler will be used )
